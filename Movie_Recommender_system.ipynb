Popularity Based Recommendation System

```python
ratings
```

```python
books
```

```python
ratings_with_name=ratings.merge(books,on="ISBN")
```

```python
ratings_with_name
```

```python
num_rating_df=ratings_with_name.groupby("Book-Title").count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_rating_df
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace = True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=100].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df.head()
```

```python
popular_df.shape
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

Collaborative Filtering

```python
ratings_with_name
```

```python
x=ratings_with_name.groupby("User-ID").count()['Book-Rating']>200
index_users=x[x].index
```

```python
ratings_with_name=ratings_with_name[ratings_with_name['User-ID'].isin(index_users)]
```

```python
filtered_rating=ratings_with_name
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
famous_books
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
y.shape
# famous_books=y[y].index
# famous_books
```

```python
famous_books=y[y].index
famous_books
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating.shape
```

```python
final_rating.duplicated().sum()
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.shape
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Calculating the similarity score of the book with other books
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
# Function to get the book recommendation
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
recommend('2nd Chance')
```

```python
recommend('Message in a Bottle')
```
